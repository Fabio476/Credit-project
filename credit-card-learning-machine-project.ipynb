```python
#this instals packagest that are going to be helpful building this machine.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Data was cleaned in in sql and then uploaded into kaggle in order to use.
data= pd.read_csv('/kaggle/input/d/fabiobarreto1/c-credit-project/Clean_credit_data.csv')
data
```

```python
#gives us the name in, non-null count, and the types of each columns.
# running the unique functions to double check that there aren't any duplicates in the file.
data.info()
data['ID'].nunique()
```

```python
#This codes checks and returns all the columns that are objects type.
ot = pd.DataFrame(data.dtypes =='object').reset_index()
object_type = ot[ot[0] == True]['index']
object_type
```

```python
# Data is clean and almost ready to be use. The only thing that needs to done before building
#the learning machine is the columns that are object to interger(7). This code tells us the
#count of each column.
a = data[object_type]['CODE_GENDER'].value_counts()
b = data[object_type]['FLAG_OWN_CAR'].value_counts()
c = data[object_type]['FLAG_OWN_REALTY'].value_counts()
d = data[object_type]['NAME_INCOME_TYPE'].value_counts()
e = data[object_type]['NAME_EDUCATION_TYPE'].value_counts()
f = data[object_type]['NAME_FAMILY_STATUS'].value_counts()
g = data[object_type]['NAME_HOUSING_TYPE'].value_counts()

print('\n',a,'\n\n',b,'\n\n',c,'\n\n',d,'\n\n',e,'\n\n',f,'\n\n',g,)
```

```python
#label encoder will help us convert the values in the object columns into numeric values.
#this is important because the learning machine needs all the information to be numeric.
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for x in data:
    if data[x].dtypes == 'object':
        data[x] = le.fit_transform(data[x])
```

```python
#All data is in numeric value.
data.info()
```

```python

X = data.iloc[:,1:-1] # all values without headers. 
y = data.iloc[:,-1] # all headers.
```

```python
#this code splits the data into two parts. 70% of the data is use to train the machine and
#the remaining 30% is use to check the accuracy of the prediction. 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

```python
y_test
```

```python
# In this data set, there is an issue of oversampling. In order to fix this issue, we use the
#smote tool in order to create sample data to conter the oversampling.  
from imblearn.over_sampling import SMOTE

oversample = SMOTE()
X_balanced, y_balanced = oversample.fit_resample(X_train, y_train)
X_test_balanced, y_test_balanced = oversample.fit_resample(X_test, y_test)
```

```python
# Checks the that the oversampling issue has been fix.
a = y_train.value_counts()
b = y_balanced.value_counts()
print(a, "\n\n", b)
```

```python
# this are the algerithem that i'm going to be using to creating and testing the learning machine.
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
classifiers = {
    "LogisticRegression" : LogisticRegression(),
    "DecisionTree" : DecisionTreeClassifier(),
    "RandomForest" : RandomForestClassifier(),
    "XGBoost" : XGBClassifier()
}
```

```python
#Test the data and provides accuracy score for all the machine learning algerithems.
train_scores = []
test_scores = []

for key, classifier in classifiers.items():
    classifier.fit(X_balanced, y_balanced)
    train_score = classifier.score(X_balanced, y_balanced)
    train_scores.append(train_score)
    test_score = classifier.score(X_test_balanced, y_test_balanced)
    test_scores.append(test_score)

print(train_scores)
print(test_scores)
```

The XGBC classifier seemed to have worked the best with an accuracy of 76%.

```python
# this will let us to set XGBC boost as the method of training and testing the machine. 
#it will provide a prediction of the scores.
xgb = XGBClassifier()
model = xgb.fit(X_balanced, y_balanced)
prediction = xgb.predict(X_test_balanced)
```

```python
# provides presentation of the scores that were produce using the XGBC boost.
from sklearn.metrics import classification_report
print(classification_report(y_test_balanced, prediction))
```
